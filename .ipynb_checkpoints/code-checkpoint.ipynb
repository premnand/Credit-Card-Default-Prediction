# Credit Card Default Predictions

## Introduction
The Credit Card Default Prediction Project aims to identify customers who are likely to default on their credit card payments in the upcoming month. Financial institutions face significant risks due to customer defaults, making it crucial to proactively identify high-risk customers to mitigate potential losses.

In this project:

We leverage historical customer data, including demographic details, credit limits, past payment behavior, and transaction records, to build a predictive model.
The process begins with data cleaning and exploratory data analysis (EDA) to uncover patterns and insights about customer behavior.
Various machine learning models, such as Logistic Regression, Decision Trees, and Random Forests, are trained and evaluated to classify customers into defaulters and non-defaulters.
The model with the best performance is fine-tuned and validated using metrics like accuracy, precision, recall, and ROC-AUC score.
The project concludes by documenting key findings and potential real-world applications, providing actionable insights for risk management in the financial sector.
This project demonstrates how data-driven decisions can help financial institutions improve credit risk management and maintain financial stability.

## Feature Engineering

### Handling Missing Values

"Although the current dataset has no missing values, a pipeline has been designed to handle missing data if it arises in future versions of the dataset. For numerical columns, missing values will be imputed using the median, and for categorical columns, the mode will be used."

```python
# Check for missing values
if df.isnull().sum().sum() > 0:
    # Handling missing values
    # Example: Fill numerical columns with the median and categorical columns with the mode
    for col in df.columns:
        if df[col].dtype in ['int64', 'float64']:
            df[col].fillna(df[col].median(), inplace=True)
        else:
            df[col].fillna(df[col].mode()[0], inplace=True)
else:
    print("No missing values detected.")
```
